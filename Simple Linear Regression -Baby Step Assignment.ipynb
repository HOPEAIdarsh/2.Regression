```python
dataset = pd.read_csv("Salary_Data.csv") #function read-csv calls under pd class then it passes salary_Data.csv file to the variable dataset
```

```python
dataset #here it will load all the datas present @ dataset
```

```python
independent=dataset[["YearsExperience"]] #assign a single column from table with there respective variable.
independent # check the variable (i.e)load the datas inside the independent variable
```

```python
dependent=dataset[["Salary"]] #assign a single column from table with there respective variable.
dependent # check the variable (i.e)load the datas inside the dependent variable
```

```python
# SPLIT TRAIN SET N TEST SET.
from sklearn.model_selection import train_test_split #in sklearn lib,under model_selection file it calls  train_test_split function.
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0) #call the imported function with its respective parameters along with train n test sets i/p and o/p variable.
```

```python
!pip install scikit-learn
```

```python
#HOW TO PASS INPUT / LEARNING PHASE
from sklearn.linear_model import LinearRegression #through sklearn lib import linear regression file inside linear_model function.
regressor = LinearRegression() #class assigned to regressor function
regressor.fit(X_train,y_train) #calls fit function using access operator,here the values are substituted ,model created.
```

```python
weight=regressor.coef_ #used to see weight value.
weight
```

```python
bias=regressor.intercept_ #used to see bias value.
bias
```

```python
#EVALUTING THE MODEL
y_Pred=regressor.predict(X_test) #predict data from test
from sklearn.metrics import r2_score #actual and predicted input passed to r2 
r_score = r2_score(y_test,y_Pred) #r2 value evaluted nearly to 1 good model else bad model.
r_score #load the value.
```

```python
# HOW TO SAVE THE MODEL
import pickle #pickle lib used to save model
filename = "finalized_model_linear.sav" #filename created for finalized model
pickle.dump(regressor,open(filename,'wb')) #model assigned to pickle.
```

```python
#CHECK THE SAVED MODEL WHETHER IT IS WORKING PROPERLY OR NOT
loaded_model = pickle.load(open("finalized_model_linear.sav",'rb')) #load the saved model(i.e)file name just read the file.
result = loaded_model.predict([[13]]) #test with given input.
result # load the value related to given input.
```
